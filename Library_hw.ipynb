вопрос 3: Какие три жанра игр самые популярные в Северной Америке (по количеству продаж)? 

```python
# ваш код здесь
df_group_genre = df.groupby(['Genre']).sum(["NA_Sales"])
df_group_genre.head(10)
df_sorted = df_group_genre.sort_values(['NA_Sales'], ascending = False)[:3]
df_sorted
```

вопрос 4: Какая платформа была самой популярной (по количеству выпущенных игр) в 2000 году? А в 2015?

```python
# ваш код здесь

# отбрасываем лишние колонки и создаем копию датасета
df_clean = df[['Name', 'Platform', 'Year']]
df_clean = df_clean.copy()

# сортируем датасет по году и платформе и считаем количество выпущенных игр на каждой платформе каждого года.
# Также выводим сумму игр в колонку с названием count, чтобы можно было к ней обратиться
df_platform_data = df_clean.groupby(['Year', 'Platform'])['Name'].count().reset_index(name = 'count')
df_platform_data.iloc[-50:]

# вводим перемнную для поиска
var_year = 2015

# находим нужную нам строку по переменной и выводим (одну) самую максимальную по значению выпущенных игр
df_top_platform = df_platform_data[df_platform_data['Year'] == var_year].nlargest(1, 'count')
print(df_top_platform)
```

вопрос 5: Какой издатель выпустил больше всего игр в период 2012-2015 оба конца включительно?

```python
# ваш код здесь

# оставляем в таблице только интервал 2012-2015 года
df_clean_year = df[df['Year'].between(2012, 2015)]
df_clean_year = df_clean_year.copy()
df_clean_year.head(20)

#группируем по издательству и считаем количество игр
df_group_publisher = df_clean_year.groupby(['Publisher'])['Name'].count().reset_index(name = 'count')
df_group_publisher.head(10)

# сортируем по количеству игр и выводим топ 1
df_top_publisher = df_group_publisher.sort_values(['count'], ascending = False)[:1]
df_top_publisher.head()
```

вопрос 6: Какой процент игр в жанре спорт был продан в Европе?

```python
# ваш код здесь

# отбрасываем лишние колонки и создаем копию датасета
df_clean = df[['Rank', 'Genre', 'EU_Sales', 'Global_Sales']]
df_clean = df_clean.copy()

# сгруппируем по жанру и просуммируем продажи 
df_only_genre = df_clean.groupby(by=['Genre']).sum()

# посчитаем процент продаж игр в Европе от всех продаж для каждого жанра
df_only_genre['percent_sales'] = df_only_genre['EU_Sales']/df_only_genre['Global_Sales']*100

# выведим строку в жанре спорт
display(df_only_genre.loc[['Sports'], ['percent_sales']])
```
